```python
import gym

from stable_baselines.common.policies import MlpPolicy
from stable_baselines.common.vec_env import DummyVecEnv
from stable_baselines import PPO2
from stable_baselines.bench import Monitor
import tensorflow as tf

#env = gym.make('Humanoid-v2')
# Optional: PPO2 requires a vectorized environment to run
# the env is now wrapped automatically when passing it to the constructor
log_dir = "./temp"
env = DummyVecEnv([lambda: Monitor(gym.make('Humanoid-v2'),log_dir, allow_early_resets=True)])

#env = DummyVecEnv(Monitor([lambda: env, "/tmp/sac/"]))

#model = PPO2(MlpPolicy, env, n_steps=4096, nminibatches=64, lam=0.95, gamma=0.99, noptepochs=10, ent_coef=0.0, learning_rate=lambda f: 1e-4 * f, cliprange= 0.1, verbose=2)
policy_kwargs = dict(act_fun=tf.nn.tanh, net_arch=[32, 32])
model = PPO2(MlpPolicy, env, verbose=1, policy_kwargs=policy_kwargs)
timesteps=50000
model.learn(total_timesteps=timesteps)

obs = env.reset()
try:
    for i in range(1000):
        action, _states = model.predict(obs)
        obs, rewards, dones, info = env.step(action)
        env.render()
finally:
    env.close()
```
